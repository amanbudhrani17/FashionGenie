```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
def getQLink(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    tp = soup.find_all('div', class_='feed-item-title clearfix')
    li=[]
    for item in tp:
        href_link = item.find('a')['href'][2:]
        li.append(href_link)
    return li
```

```python
def getAnswers(main_url):
    response = requests.get(main_url)
    soup = BeautifulSoup(response.content, "html.parser")
    tp = soup.find_all('div', class_="user-content clearfix")
    li = []
    for i in tp:
        vt = i.get_text(strip=True)
        li.append(vt)
    return li
```

```python
def getQuestionLinks():
    url = 'https://www.blurtit.com/Fashion/?page='
    d = {}
    for i in range(1,55):
        u = url+ str(i)
        response = requests.get(u)
        soup = BeautifulSoup(response.content, "html.parser")
        tp = soup.find_all('div', class_='feed-item-title clearfix')
        a=0
        for item in tp:
            qa = item.get_text(strip=True)
            href_link = item.find('a')['href']
            d[qa] = getAnswers('https:'+href_link)
            print(a)
            a=a+1
        print(" ")
        print(i)
        print(" ")
        
    return d
            
```

```python
d3={}
for key in d:
    s = key.lower()
    if 'where' in s:
        continue
    else:
        d3[key] = d[key]
```

```python
json_data = []
str1 = "question"
str2 = "answer"
for question, answers in d3.items():
    d10 = {}
    d10[str1] = question
    str3=""
    for i in answers:
        str3=str3+" "+i
    d10[str2] = str3[0:min(300,len(str3))]
    json_data.append(d10)

# Save JSON object to a file
with open('output.json', 'w') as json_file:
    json.dump(json_data, json_file, indent=4)

print("JSON file 'output.json' created.")
```
